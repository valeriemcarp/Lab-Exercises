```python
source = requests.get("https://www.imdb.com/list/ls055592025/").text
soup = BeautifulSoup(source, "lxml")
```

```python
#Top 100 Movies Web Scraping:
for element in soup.find_all("div", class_="lister-item-content"):
    title = element.a.text
    date = element.find("span", class_="lister-item-year text-muted unbold").text.strip()
    rating = element.find("span", class_="certificate").text.strip()
    print(title)
    print (date)
    print(rating)
```

```python
#Top 100 Movies CSV:
new_csv = open("Top100Movies.csv", "w", newline = "", encoding = "utf-8")
csv_writer = csv.writer(new_csv)
csv_writer.writerow(["Title", "Date", "Rating"])

for element in soup.find_all("div", class_="lister-item-content"):
    title = element.a.text
    date = element.find("span", class_="lister-item-year text-muted unbold").text.strip()
    rating = element.find("span", class_="certificate").text.strip()
    
    csv_writer.writerow([title, date, rating])
    
new_csv.close()

sheet = pd.read_csv("Top100Movies.csv")
sheet
```

```python
#First 5 Pages (Title, Description, & Hyperlink) Scraped:
page = 1
site = "https://www.loc.gov/search/?q=cats&sp=1"
want = requests.get(site)
content = want.text
soup = BeautifulSoup(content, "html.parser")

links = []
while page != 6:
    library = requests.get(f"https://www.loc.gov/search/?q=cats&sp={page}").text
    soup = BeautifulSoup(library, "lxml")
    
    for element in soup.find_all("span", class_="item-description-title"):
        title = element.text.strip()
        print(f"{title}")
        
    for element in soup.find_all("span", class_="item-description-abstract"):
        description = element.text.strip()
        print(f"{description}")
        
    for h in soup.find_all("figure"):
        a = h.find("a")
    try:
        if "href" in a.attrs: 
            site = a.get("href")
            links.append(site)
    except:
        pass
    
    for site in links:
        print(site)
    
    page += 1
```

```python
#First 5 Pages Scraped in a CSV:
cat_csv = open("Cat_Gov_Search.csv", "w", newline = "", encoding = "utf-8")
cat_writer = csv.writer(cat_csv)
cat_writer.writerow(["title", "description", "links"])

page = 1
site = "https://www.loc.gov/search/?q=cats&sp=1"
want = requests.get(site)
content = want.text
soup = BeautifulSoup(content, "html.parser")

links = []
while page != 6:
    library = requests.get(f"https://www.loc.gov/search/?q=cats&sp={page}").text
    soup = BeautifulSoup(library, "lxml")
    
    for element in soup.find_all("span", class_="item-description-title"):
        title = element.text.strip()
        print(f"{title}")
        
    for element in soup.find_all("span", class_="item-description-abstract"):
        description = element.text.strip()
        print(f"{description}")
        
    for h in soup.find_all("figure"):
        a = h.find("a")
    try:
        if "href" in a.attrs: 
            site = a.get("href")
            links.append(site)
    except:
        pass
    
    for site in links:
        print(site)
    
    cat_writer.writerow([title, description, links])
    
    page += 1
    
sheet = pd.read_csv("Cat_Gov_Search.csv")
sheet
```
